```python
import pandas as pd

df = pd.read_excel("job_listings.xlsx")

print(df.head())
print(df.shape)
```

```python
df.drop_duplicates(inplace=True)

print("Duplicates removed")
```

```python
df['salary'].fillna("Not Disclosed", inplace=True)
```

```python
df['company'] = df['company'].str.title()
```

```python
df['search_role'] = "Data Engineer"
```

```python
print(df.head())
print(df.info())
```

```python
import sqlite3

# Create database connection
conn = sqlite3.connect("jobs.db")

# Load dataframe into SQL table
df.to_sql(
    "job_listings",
    conn,
    if_exists="replace",
    index=False
)

print("Data loaded into SQLite database successfully!")
```

```python
query = "SELECT * FROM job_listings LIMIT 10"

result = pd.read_sql(query, conn)

print(result)
```

```python
conn.close()
```
