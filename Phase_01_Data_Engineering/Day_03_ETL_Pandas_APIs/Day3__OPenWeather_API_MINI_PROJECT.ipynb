```python
import pandas as pd
```

```python
weather_records = [
    {
        "city": "Chennai",
        "country": "India",
        "temperature": 34,
        "feels_like": 38,
        "humidity": 70,
        "pressure": 1008,
        "weather": "clouds",
        "description": "broken clouds",
        "wind_speed": 4.5
    },
    {
        "city": "Bangalore",
        "country": "India",
        "temperature": 28,
        "feels_like": 30,
        "humidity": 65,
        "pressure": 1010,
        "weather": "rain",
        "description": "light rain",
        "wind_speed": 3.2
    },
    {
        "city": "Mumbai",
        "country": "India",
        "temperature": 32,
        "feels_like": 36,
        "humidity": 75,
        "pressure": 1006,
        "weather": "haze",
        "description": "moderate haze",
        "wind_speed": 5.1
    },
    {
        "city": "Delhi",
        "country": "India",
        "temperature": 39,
        "feels_like": 42,
        "humidity": 45,
        "pressure": 1002,
        "weather": "clear",
        "description": "clear sky",
        "wind_speed": 2.8
    }
]

print("Weather data extracted successfully")
print("Total records:", len(weather_records))
```

```python
df = pd.DataFrame(weather_records)

print("Original Weather Data:")
print(df)
```

```python
df["city"] = df["city"].str.title()
df["country"] = df["country"].str.title()
df["weather"] = df["weather"].str.title()
df["description"] = df["description"].str.title()

print("Text columns cleaned successfully")
```

```python
df.drop_duplicates(inplace=True)

print("Duplicate records removed")
```

```python
df["temperature_fahrenheit"] = (df["temperature"] * 9/5) + 32

print("Temperature Fahrenheit column added")
```

```python
df.to_csv("cleaned_weather_data.csv", index=False)

print("Cleaned weather data saved as CSV file")
```
