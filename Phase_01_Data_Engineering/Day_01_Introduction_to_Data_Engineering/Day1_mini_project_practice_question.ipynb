```python
import pandas as pd

# -------------------------------
# Student Data Explorer
# -------------------------------

data = {
    "name": ["Asha", "Ravi", "Meena", "Karthik", "Priya", "Arun", "Divya", "Sanjay"],
    "gender": ["Female", "Male", "Female", "Male", "Female", "Male", "Female", "Male"],
    "department": ["CSE", "Electronics", "IT", "Electronics", "CSE", "IT", "Electronics", "CSE"],
    "math_score": [85, 76, 92, 68, 88, 74, 95, 60],
    "programming_score": [90, 70, 95, 65, 85, 72, 98, 55],
    "attendance_percentage": [92, 80, 96, 70, 89, 73, 94, 78]
}

df = pd.DataFrame(data)

# Creating total_score column
df["total_score"] = df["math_score"] + df["programming_score"]

# 1. Project Title
print("=" * 50)
print("        STUDENT PERFORMANCE EXPLORER")
print("=" * 50)

# 2. Dataset Overview
print("\nDATASET OVERVIEW")
print("-" * 50)
print(f"Total number of students: {len(df)}")
print(f"Total number of columns: {df.shape[1]}")
print(f"Number of unique departments: {df['department'].nunique()}")

# 3. Department wise Analysis
print("\nDEPARTMENT WISE ANALYSIS")
print("-" * 50)
print(df["department"].value_counts())

# 4. Score Analysis
print("\nSCORE ANALYSIS")
print("-" * 50)
print(f"Highest Math Score: {df['math_score'].max()}")
print(f"Lowest Math Score: {df['math_score'].min()}")
print(f"Average Math Score: {df['math_score'].mean():.2f}")
print(f"Average Programming Score: {df['programming_score'].mean():.2f}")
print(f"Average Attendance Percentage: {df['attendance_percentage'].mean():.2f}%")

# 5. Gender wise Analysis
print("\nGENDER WISE ANALYSIS")
print("-" * 50)
print(df.groupby("gender")["total_score"].mean())

# 6. Top Performers
print("\nTOP 5 PERFORMERS")
print("-" * 50)
top_5 = df.sort_values(by="total_score", ascending=False).head(5)

for index, row in top_5.iterrows():
    print(f"Name: {row['name']}, Department: {row['department']}, Total Score: {row['total_score']}")

# 7. Attendance Analysis
print("\nATTENDANCE ANALYSIS")
print("-" * 50)

low_attendance = df[df["attendance_percentage"] < 75]

if low_attendance.empty:
    print("All students have attendance above 75%.")
else:
    for index, row in low_attendance.iterrows():
        print(f"Name: {row['name']}, Department: {row['department']}, Attendance: {row['attendance_percentage']}%")

# -------------------------------
# Practice Questions
# -------------------------------

print("\nPRACTICE QUESTIONS ANSWERS")
print("-" * 50)

# 1. Electronics students count
electronics_count = df[df["department"] == "Electronics"].shape[0]
print(f"1. Students from Electronics Department: {electronics_count}")

# 2. Average attendance of female students
female_avg_attendance = df[df["gender"] == "Female"]["attendance_percentage"].mean()
print(f"2. Average attendance of female students: {female_avg_attendance:.2f}%")

# 3. Student with lowest programming score
lowest_prog_student = df.loc[df["programming_score"].idxmin()]
print(f"3. Student with lowest programming score: {lowest_prog_student['name']}")

# 4. Students scored above 80 in math
math_above_80 = df[df["math_score"] > 80].shape[0]
print(f"4. Students scored above 80 in Math: {math_above_80}")

# 5. Average total_score for attendance above 90%
avg_total_above_90 = df[df["attendance_percentage"] > 90]["total_score"].mean()
print(f"5. Average total_score for attendance above 90%: {avg_total_above_90:.2f}")
```
