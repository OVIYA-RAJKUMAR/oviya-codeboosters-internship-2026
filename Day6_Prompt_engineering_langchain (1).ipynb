```python
import os
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
os.environ["GROQ_API_KEY"]="----------------------------------"
llm=ChatGroq(model="llama-3.3-70b-versatile",temperature=0)
prompt=PromptTemplate(
    input_variables=["instruction"],
    template="Follow the instruction below.\n{instruction}"
)

parser=StrOutputParser()
chain=prompt|llm|parser
response=chain.invoke({"instruction":"who won ipl 2023?"})
print(response)
```

```python
import os

from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
os.environ["GROQ_API_KEY"] = "--------------------------------"
llm = ChatGroq(model="llama-3.3-70b-versatile",temperature=0)
prompt = PromptTemplate(
    input_variables=["instruction"],
    template="""
Follow the examples below.
Question: What does HTML stand for?
Answer: HyperText Markup Language
Question: Which language is used for data analysis?
Answer: Python
Question: {instruction}
Answer:
"""
)
parser = StrOutputParser()
chain = prompt | llm | parser
response = chain.invoke({"instruction": "Who invented Python programming language?"})
print(response)
```
