````python
import os
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

os.environ["GROQ_API_KEY"] = "--------------------------------------------------"

llm = ChatGroq(model="llama-3.1-8b-instant")

prompt = PromptTemplate.from_template("""
You are a Streamlit code generator.

Rules:
1. Return only executable Streamlit Python code.
2. Do not give explanation.
3. Do not use markdown.
4. Do not use ```python.
5. Code must start with import streamlit as st.
6. App must have title, inputs, button, and output.
7. Code must run directly as app.py.

User request:
{question}

Streamlit code:
""")

agent = prompt | llm | StrOutputParser()

question = input("Enter the Streamlit app you want: ")

generated_code = agent.invoke({"question": question})

print(generated_code)
````

````python
streamlit_code = """
import streamlit as st
import operator

st.title("Calculator")

col1, col2 = st.columns(2)

with col1:
    num1 = st.number_input("Number 1")

with col2:
    num2 = st.number_input("Number 2")

op = st.selectbox("Operation", ["Addition (+)", "Subtraction (-)", "Multiplication (*)", "Division (/)"])

if st.button("Calculate"):
    if op == "Addition (+)":
        result = operator.add(float(num1), float(num2))
    elif op == "Subtraction (-)":
        result = operator.sub(float(num1), float(num2))
    elif op == "Multiplication (*)":
        result = operator.mul(float(num1), float(num2))
    elif op == "Division (/)":
        if num2 != 0:
            result = operator.truediv(float(num1), float(num2))
        else:
            result = "Error: Division by zero"
    else:
        result = "Error: Invalid operation"

    st.write(result)

"""

with open("app.py", "w", encoding="utf-8") as file:
    file.write(streamlit_code)

print("app.py created successfully!")
````
