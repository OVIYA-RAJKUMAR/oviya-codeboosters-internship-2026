```python
!pip install llama-index sentence-transformers -q

import os
from textwrap import dedent
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex, Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

os.makedirs("data", exist_ok=True)

file_path = "data/sece_aids_Dept.txt"

content = dedent("""
Centre of Excellence - Data Science

Data Science is instrumental in extracting valuable insights from vast datasets.

ABOUT CoE
Welcome to the Centre for Data Science.

OBJECTIVES
Provide training programs in Data Science.
Conduct hands-on projects and hackathons.
Support research and innovation.

CENTRE HEAD
Ms. Parvathy K, Assistant Professor/AIDS

CENTRE COORDINATORS
Ms.G.Durga Devi, Assistant Professor/AIDS

EVENTS
1 Mastering in MLOps
2 Introduction to Data Bricks
3 Fundamentals of Generative AI
4 Hands-on Session on Power BI

CoE OFFICE
Sri Eshwar College of Engineering
Coimbatore
""")

with open(file_path, "w", encoding="utf-8") as file:
    file.write(content)

print("Text file created successfully!")

documents = SimpleDirectoryReader("data").load_data()

print("Documents loaded successfully!")

Settings.embed_model = HuggingFaceEmbedding(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

index = VectorStoreIndex.from_documents(documents)

retriever = index.as_retriever(similarity_top_k=3)

query = "Who is the centre head?"

results = retriever.retrieve(query)

print("\nQuery Results:\n")

for result in results:
    print(result.text)
    print("-" * 50)
```

```python
import os
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex, Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.llms.groq import Groq
os.environ['GROQ_API_KEY']="-----------------------------------------------------------------------"
llm=Groq(model="llama-3.1-8b-instant")
embed_model = HuggingFaceEmbedding(model_name="sentence-transformers/all-MiniLM-L6-v2")
file_path = "data/sece_aids_Dept.txt"
documents=SimpleDirectoryReader("data").load_data()
index = VectorStoreIndex.from_documents(documents,llm=llm,embed_model=embed_model)
query_engine = index.as_query_engine(llm=llm)
response=query_engine.query("Who is the centre head?")
print(response)
```
