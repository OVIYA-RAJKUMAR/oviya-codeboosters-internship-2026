```python
from google.colab import userdata
import os, base64, requests

GITHUB_TOKEN = userdata.get("GITHUB_PAT")

OWNER = "OVIYA-RAJKUMAR"
REPO = "PRACTICE"
BRANCH = "main"
```

```python
FILES_TO_PUSH = [
    "/content/index.html"
]
```

```python
url = f"https://api.github.com/repos/{OWNER}/{REPO}"

headers = {
    "Authorization": f"token {GITHUB_TOKEN}",
    "Accept": "application/vnd.github+json"
}

res = requests.get(url, headers=headers)

print(res.status_code)
print(res.text)
```

```python
def push_file(local_path):
    file_name = os.path.basename(local_path)

    with open(local_path, "rb") as f:
        content = base64.b64encode(f.read()).decode()

    url = f"https://api.github.com/repos/{OWNER}/{REPO}/contents/{file_name}"

    headers = {
        "Authorization": f"Bearer {GITHUB_TOKEN}",
        "Accept": "application/vnd.github+json"
    }

    get_res = requests.get(url, headers=headers, params={"ref": BRANCH})

    data = {
        "message": f"Auto push {file_name} from Colab",
        "content": content,
        "branch": BRANCH
    }

    if get_res.status_code == 200:
        data["sha"] = get_res.json()["sha"]

    res = requests.put(url, headers=headers, json=data)

    if res.status_code in [200, 201]:
        print(f"✅ Successfully pushed: {file_name}")
    else:
        print(f"❌ Failed to push: {file_name}")
        print(res.status_code, res.text)
```

```python
for file in FILES_TO_PUSH:
    if os.path.exists(file):
        push_file(file)
    else:
        print(f"⚠️ File not found: {file}")
```

```python
import os, base64, requests

OWNER = "OVIYA-RAJKUMAR"
REPO = "PRACTICE"
BRANCH = "main"

FILES_TO_PUSH = ["/content/index.html"]

headers = {
    "Authorization": f"token {GITHUB_TOKEN}",
    "Accept": "application/vnd.github+json"
}

def push_file(local_path):
    file_name = os.path.basename(local_path)
    url = f"https://api.github.com/repos/{OWNER}/{REPO}/contents/{file_name}"

    with open(local_path, "rb") as f:
        content = base64.b64encode(f.read()).decode()

    data = {
        "message": f"Add {file_name} from Colab",
        "content": content,
        "branch": BRANCH
    }

    res = requests.put(url, headers=headers, json=data)

    print("Status:", res.status_code)
    print(res.text)

for file in FILES_TO_PUSH:
    push_file(file)
```
