```python
from crewai import Agent, Task, Crew

def research_agent_example():

    researcher = Agent(
        role="Research Analyst",
        goal="Identify the top AI trends in 2025",
        backstory="""
        You are an experienced AI research analyst.
        You specialize in tracking emerging technologies,
        industry trends, and innovations in Artificial Intelligence.
        """,
        verbose=True,
        llm="groq/llama-3.3-70b-versatile"
    )

    research_task = Task(
        description="List and analyze the top 3 AI trends in 2025 with examples.",
        expected_output="Comprehensive report with detailed explanation for each trend.",
        agent=researcher
    )

    crew = Crew(
        agents=[researcher],
        tasks=[research_task],
        verbose=True
    )

    result = crew.kickoff()
    result=research_agent_example()
    return result
```
